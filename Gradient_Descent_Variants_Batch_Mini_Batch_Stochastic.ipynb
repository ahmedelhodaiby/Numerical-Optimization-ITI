#### Import numpy, matplotlib.pyplot and make it inline

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score

%matplotlib inline
```

To have a dataset or set of data points, the student must generate a pair of arrays <b>X</b> and <b>y</b> with the values in <b>X</b> equally distributed between <b>0</b> and <b>20</b> and the values in <b>y</b> such that: 
<b>yi = a*xi + b (and a = -2, b = 1)</b>


```python
x = np.linspace(0, 20)
y = -2 * x + 1
```

#### Use your code from practical session 1 and create a function to perform GD for single variable linear regression.
#### The function should take the following input parameters:
##### Input data (X), Target Labels (y), Learning Rate (alpha), Maximum number of iterations
#### The function should return the following:
##### All Theta_0 and Theta_1 values till converge, and equivellant Losses, all hypothesis output, and optimum values of Theta_0 and Theta_1
#### Note: you can choose any suitable stop criteria.

```python
def gradientDescentSingleVar(x_data, y_data, alpha, max_iterations):
    theta0 = 0
    theta1 = 0
    costList = []
    theta0_list = []
    theta1_list = []
    hypothesis_list = []
    for i in range(max_iterations):
        hypothesis = theta0 + theta1 * x_data
        hypothesis_list.append(hypothesis)
        error = hypothesis - y_data
        cost = np.sum(((error)**2))/len(x_data)
        costList.append(cost)
        gradient_theta0 = (1/len(x_data))*sum(error)
        gradient_theta1 = (1/len(x_data))*sum(error * x_data)
        theta0 = theta0 - alpha * gradient_theta0
        theta0_list.append(theta0)
        theta1 = theta1 - alpha * gradient_theta1
        theta1_list.append(theta1)
        if((gradient_theta0**2 + gradient_theta1**2) < 1e-3):
            print("it took {} iterations".format(i))
            break
    return theta0_list, theta1_list, costList, hypothesis_list, theta0, theta1
```

#### Try your function with any input parameterss of your choice.

```python
theta0_list, theta1_list, costList, hypothesis_list, theta0, theta1 = gradientDescentSingleVar(x, y, 0.011,1000)
```

#### Update the implementation of the Batch GD to be Mini-Batch and repeat the above plots and evaluations
#### Change batch size and see the effect

```python
def gradientDescentMiniBatch(x_data, y_data, alpha, max_iterations, batchsize):
    theta0 = 0
    theta1 = 0
    costList = []
    theta0_list = []
    theta1_list = []
    hypothesis_list = []
    for i in range(max_iterations):
        for j in range(0, len(x_data), batchsize):

            x_batch = x_data[j : j+batchsize]
            y_batch = y_data[j : j+batchsize]

            hypothesis = theta0 + theta1 * x_batch
            hypothesis_list.append(hypothesis)

            error = hypothesis - y_batch
            cost = np.sum(((error)**2))/len(x_batch)
            costList.append(cost)

            gradient_theta0 = (1/len(x_batch))*sum(error)
            gradient_theta1 = (1/len(x_batch))*sum(error * x_batch)

            theta0 = theta0 - alpha * gradient_theta0
            theta0_list.append(theta0)
            theta1 = theta1 - alpha * gradient_theta1
            theta1_list.append(theta1)
            
    return theta0_list, theta1_list, costList, hypothesis_list, theta0, theta1
```

```python
theta0_list_mini, theta1_list_mini, costList_mini, hypothesis_mini, theta0_mini, theta1_mini = gradientDescentMiniBatch(x, y, 0.001,1000, 5)
```

```python
y_predict = theta0_mini +theta1_mini * x
r2_score(y, y_predict)
```

```python
fig = plt.figure(figsize=(20, 10))

axis1 = fig.add_subplot(231)
plt.scatter(x, y)
yplot = theta0_list_mini[-1] + theta1_list_mini[-1] * x
axis1.plot(x, yplot, 'g')

axis2 = fig.add_subplot(232)
for i in range(len(theta0_list_mini)):
    ypoint = theta0_list_mini[i] + theta1_list_mini[i] * x
    axis2.plot(x, ypoint)

axis3 = fig.add_subplot(233)
axis3.plot(theta0_list_mini, costList_mini)
plt.xlabel("theta 0")
plt.ylabel("loss")

axis4 = fig.add_subplot(234)
plt.plot(theta1_list_mini, costList_mini)
plt.xlabel("theta 1")
plt.ylabel("loss")

axis5 = fig.add_subplot(235)
plt.plot(costList_mini)
plt.xlabel("iterations")
plt.ylabel("loss")

fig.suptitle("Mini batch Linear Regresion at alpha=0.001 implemented with 5 mini batches")
axis1.set_title("Best regression line")
axis2.set_title("All regression lines till converge")
axis3.set_title("Loss vs. Theta 1")
axis4.set_title("Loss vs. Theta 1")
axis5.set_title("Loss vs. Epochs")

plt.show()
```

## Stochastic GD

#### Update the implementation of the Batch GD to beStochastic and repeat the above plots and evaluations
#### You can try different learning rate and see its effect.
#### Compare to Batch and Mini-Batch GD

```python
def gradientDescentStochastic(x_data, y_data, alpha, max_iterations):
    theta0 = 0
    theta1 = 0
    costList = []
    theta0_list = []
    theta1_list = []
    hypothesis_list = []
    for i in range(max_iterations):
        for j in range(0, len(x_data)):
            hypothesis = theta0 + theta1 * x_data[j]
            hypothesis_list.append(hypothesis)

            error = hypothesis - y_data[j]
            cost = (error)**2
            costList.append(cost)

            gradient_theta0 = error * 2
            gradient_theta1 = (error * x_data[j]) * 2

            theta0 = theta0 - alpha * gradient_theta0
            theta0_list.append(theta0)
            theta1 = theta1 - alpha * gradient_theta1
            theta1_list.append(theta1)
            
    return theta0_list, theta1_list, costList, hypothesis_list, theta0, theta1
```
